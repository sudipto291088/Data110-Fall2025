# 1. Introduction

The dataset titled “Data Scientist Jobs in 2021” contains information about job listings for data-related roles, including job titles, locations, salaries, company details, industry sectors, and technical skill requirements.

I chose this dataset because it directly aligns with my career focus in Data Science and AI, offering practical insights into what the job market looks like — salary trends, demand by industry, and which skills or degrees correlate with higher pay.

This project’s goal was to understand the structure of the data, perform exploratory data analysis (EDA), engineer meaningful features, and model salary prediction.

# 2. Summary of the EDA Process
Step 1–2: Data Loading and Cleaning

Imported the dataset into Pandas as df.

Detected and dropped unnecessary or redundant columns such as Type of ownership, competitors, and Employer Provided.

Replaced unwanted placeholders like -1, na, and null with NaN or meaningful categories (e.g., 'Not Specified' for missing degrees).

Created a cleaned version named df_clean for all further analysis.

Step 3: Univariate EDA

Analyzed the distribution of salaries, ratings, company ages, and categorical features (degree, industry, sector).

Key observations:

Salaries are right-skewed, with most jobs offering mid-range pay.

Degree column showed that around 20% of job postings did not specify any educational requirement.

Company ratings clustered around 3.5–4.0, indicating a moderately positive employer reputation.


###############################################################################################


Step 4–5: Bivariate EDA

Explored relationships between salary and other factors.

Heatmap: revealed strong correlations among Lower Salary, Upper Salary, and Avg Salary(K) — confirming consistency in data.

Scatterplot (Rating vs Salary): weak positive correlation — higher-rated companies tend to pay slightly more.

Boxplots:

Job Title vs Salary showed Senior and ML roles earning higher medians.

Degree vs Salary indicated mild upward trends for higher degrees.

Barplots (Industry/Location vs Salary):

Top-paying industries: Tech, Finance, Biotechnology.

Top-paying locations: San Francisco, Seattle, and New York.

Step 6: Ranking Summaries

##Step 8: Outlier Analysis

Boxplots and z-score detection confirmed the presence of a few high-end salary outliers, which appear realistic (executive roles).

Salary distribution was right-skewed, suggesting the use of mean/median reporting.

# 3. Conclusion & Recommendations
Key Findings

Industry & Location are the dominant salary drivers.

Degree requirements have a smaller but positive impact on compensation.

Skill presence (Python, AWS, Spark, SQL) aligns with higher pay ranges.

Salaries show large variation across roles and regions — data science remains a competitive but rewarding field.

Recommendations

Future modeling could test Random Forest or Gradient Boosting to capture nonlinear salary influences.

Further research could explore time-based data (e.g., salaries by year) to track market evolution.

Adding company size and remote work variables could improve predictive strength.

Overall Summary

The EDA revealed a clear structure in the 2021 data science job market:
high-rated companies and tech-centric industries offer the best compensation,
and possessing multiple technical skills significantly improves earning potential.
