```python
from geopy.geocoders import Nominatim
import time
```

```python
##setting up the geocoder
geolocator = Nominatim(user_agent='walkability_cbsa_mapper')



## Adding Lat/Long columns
cbsa_grouped['Latitude']= None
cbsa_grouped['Longitude']=None


##Geocoding each CBSA_Name
for i, row in cbsa_grouped.iterrows():
    try:
        location = geolocator.geocode(f"{row['CBSA_Name']},USA")
        if location:
            cbsa_grouped.at[i,"Latitude"]=location.latitude
            cbsa_grouped.at[i,"Longitude"]=location.longitude
        time.sleep(1) # Prevent rate limiting
    except:
        continue
```

```python
cbsa_grouped.to_csv("cbsa_walkability_geocoded.csv", index=False)
```

```python
## Step1: Initializing base map
walk_map = folium.Map(location = [39.5,-98.35],zoom_start=4,tiles="CartoDB positron")

## Step2: Adding a circle marker for each metro
for i, row in cbsa_grouped.iterrows():
    if pd.notnull(row["Latitude"]) and pd.notnull(row['Longitude']):
        popup_text = (
            f"<b>{row['CBSA_Name']}</b><br>"
            f"Avg walkability Index: <b>{row['Avg_Walkability']:.2f}</b>"
        )
        folium.CircleMarker(
            location=[row['Latitude'],row['Longitude']],
            radius=6,
            color='blue',
            fill=True,
            popup=popup_text
                 
        ).add_to(walk_map)
```

```python
walk_map.save("walkability_cbsa_map.html")
```

```python
##Trying to add color by Walkability score

import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
norm = colors.Normalize(vmin = cbsa_grouped['Avg_Walkability'].min(),
                        vmax = cbsa_grouped['Avg_Walkability'].max()
                       )
cmap = cm.get_cmap('plasma') ## Yellow-Green_Blue colormap

## Function to map color based on walkability

def get_color(value):
    rgba = cmap(norm(value))
    return colors.to_hex(rgba)
```

```python
## Creating colored Circle Marker Map


# Base map
walk_map = folium.Map(location=[39.5, -98.35], zoom_start=4, tiles="CartoDB positron")

# Add colored markers
for i, row in cbsa_grouped.iterrows():
    if pd.notnull(row['Latitude']) and pd.notnull(row['Longitude']):
        color = get_color(row['Avg_Walkability'])
        popup_text = (
            f"<b>{row['CBSA_Name']}</b><br>"
            f"Avg Walkability Index: <b>{row['Avg_Walkability']:.2f}</b>"
        )
        folium.CircleMarker(
            location=[row['Latitude'], row['Longitude']],
            radius=6,
            color=color,
            fill=True,
            fill_color=color,
            fill_opacity=0.8,
            popup=popup_text
        ).add_to(walk_map)

# Save map
walk_map.save("walkability_cbsa_colored_map.html")
```

```python
## Moving over to 2nd Interactive Visualization
## Mapping the % of 0 car Households.
```

```python
## Grouping by CBSA and computing the average % AO0

ao0_grouped = eda_df.groupby('CBSA_Name').agg({
    'Pct_AO0': 'mean'
}).reset_index()

ao0_grouped.rename(columns={'Pct_AO0': 'Avg_Pct_AO0'}, inplace=True)
```

```python
## Merging with the Geo Coded Dataset

ao0_merged = pd.merge(cbsa_grouped, ao0_grouped, on='CBSA_Name', how='inner')
```

```python

## Creating Folium Circle Marker Map with AO0 coloring

# Normalizing 0-car household % for color gradient
norm = colors.Normalize(
    vmin=ao0_merged['Avg_Pct_AO0'].min(),
    vmax=ao0_merged['Avg_Pct_AO0'].max()
)
cmap = cm.get_cmap('OrRd')  # Orange-Red color scale

def get_color(value):
    rgba = cmap(norm(value))
    return colors.to_hex(rgba)

# Base map
ao0_map = folium.Map(location=[39.5, -98.35], zoom_start=4, tiles="CartoDB positron")

# Adding circle markers
for i, row in ao0_merged.iterrows():
    if pd.notnull(row['Latitude']) and pd.notnull(row['Longitude']):
        popup_text = (
            f"<b>{row['CBSA_Name']}</b><br>"
            f"Avg % 0-Car Households: <b>{row['Avg_Pct_AO0']:.2%}</b>"
        )
        folium.CircleMarker(
            location=[row['Latitude'], row['Longitude']],
            radius=6,
            color=get_color(row['Avg_Pct_AO0']),
            fill=True,
            fill_color=get_color(row['Avg_Pct_AO0']),
            fill_opacity=0.8,
            popup=popup_text
        ).add_to(ao0_map)
```

```python
ao0_map.save("zero_car_households_cbsa_map.html")
```

```python
## 3rd Interactive Visualization
## Interactive Choropleth Map for the Average Walkability Index by US state
## Note: Apart from the 50 US States we also consider the Federal District and the 5 major territories
```

```python
## Grouping by state and computing the mean of both Walkability Index + Pct_AO0 HH


state_grouped = eda_df.groupby('STATEFP').agg({
    'NatWalkInd': 'mean',
    'Pct_AO0': 'mean'
}).reset_index()

state_grouped.rename(columns={
    'NatWalkInd': 'Avg_Walkability',
    'Pct_AO0': 'Avg_Pct_AO0'
}, inplace=True)
```

```python
## Abbreviating the states names


fp_to_state = {
    1: 'AL', 2: 'AK', 4: 'AZ', 5: 'AR', 6: 'CA', 8: 'CO', 9: 'CT', 10: 'DE',
    11: 'DC', 12: 'FL', 13: 'GA', 15: 'HI', 16: 'ID', 17: 'IL', 18: 'IN',
    19: 'IA', 20: 'KS', 21: 'KY', 22: 'LA', 23: 'ME', 24: 'MD', 25: 'MA',
    26: 'MI', 27: 'MN', 28: 'MS', 29: 'MO', 30: 'MT', 31: 'NE', 32: 'NV',
    33: 'NH', 34: 'NJ', 35: 'NM', 36: 'NY', 37: 'NC', 38: 'ND', 39: 'OH',
    40: 'OK', 41: 'OR', 42: 'PA', 44: 'RI', 45: 'SC', 46: 'SD', 47: 'TN',
    48: 'TX', 49: 'UT', 50: 'VT', 51: 'VA', 53: 'WA', 54: 'WV', 55: 'WI',
    56: 'WY'
}

state_grouped["State"] = state_grouped["STATEFP"].map(fp_to_state)
```

```python
## Loading the US states GeoJSOn


us_states_geojson = "https://raw.githubusercontent.com/python-visualization/folium/master/examples/data/us-states.json"
```

```python

# The Base map
state_map = folium.Map(location=[39.5, -98.35], zoom_start=4, tiles="CartoDB positron")

# Adding choropleth layer
folium.Choropleth(
    geo_data=us_states_geojson,
    name="choropleth",
    data=state_grouped,
    columns=["State", "Avg_Walkability"],
    key_on="feature.id",  # Matches the 'State' abbreviation in the GeoJSON
    fill_color="YlGnBu",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="Average Walkability Index by State"
).add_to(state_map)

```

```python
## Saving the HTML 

state_map.save("walkability_choropleth_by_state.html")
```
