# The columns present in the Titanic dataset are:

Survived: This column contains information about whether a passenger survived the sinking or not. The value 0 indicates that the passenger did not survive, while the value 1 indicates that the passenger did survive.

Pclass: This column contains information about the passenger’s class. The value 1 indicates first class, the value 2 indicates second class, and the value 3 indicates third class.

Sex: This column contains information about the passenger’s gender.

Age: This column contains the age of the passenger.

SibSp: This column contains information about the passenger’s siblings and spouse.

Parch: This column contains information about the passenger’s parents and children.

Fare: This column contains the fare paid by the passenger.

class: This columns lists out the pclass in words

who: This columns indicates whether the passenger is a male, female or a child

adult_male: This columns is to indicate if the passenger was an Adult Male

deck: This columns is a categorical variable derived from the Cabin column, which indicates the deck (level) of the ship where a passenger's cabin was located. The deck is represented by a single letter (e.g., 'A', 'B', 'C', 'D', 'E', 'F', 'G'), with each letter corresponding to a different deck on the ship. This feature is used to analyze how a passenger's location might have impacted their survival chances.

embark_town: This column contains information about the port of embarkation.Cherbourg, Queenstown, Southampton.

alive: Whether the particular passenger was alive

alone: If the passenger was a solo traveller.

```python
## While trying to determine the null values we find that 'deck' and 'age' among all the other columns contain high number of missing values

df.isna().sum().sort_values(ascending=False)
```

```python
## Determining the percentage, we notice that Deck has about 77% of missing values. Since it is too high we will remove deck from the dataset

print('Determining the percent of missing values:\n ')
print(round(df.isna().sum().sort_values(ascending=False) * 100/df.shape[0]),2)
```

```python
df.dtypes
```

```python
# Drop 'deck' because ~77% missing
df = df.drop(columns=['deck'])

# Filling numeric missing values with median
df['age'] = df['age'].fillna(df['age'].median())

# Filling categorical missing values with mode
for col in ['embarked', 'embark_town']:
    df[col] = df[col].fillna(df[col].mode()[0])


df.isna().sum()
```

Task 2 — Fare distribution (histogram + KDE).
Create a histogram and a KDE of fare. Discuss right-skew, outliers, and a socioeconomic interpretation.




Task 3 — Fare × Survival (distribution comparison).
Compare fare distributions for survived = 0 vs 1 (e.g., overlaid KDEs). Focus your interpretation on distribution shapes, not just means.

```python
plt.figure(figsize=(8,5))
sns.histplot(df['fare'], bins=40, kde=True, color='teal')
plt.title('Distribution of Passenger Fares')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

# Discussion for Task 2
------


The fare distribution is strongly right-skewed: most passengers paid relatively low fares, with a long tail extending toward higher values. A few outliers appear well above £100, representing luxury-class tickets purchased by the wealthiest passengers. This skew reflects the socio-economic stratification aboard the Titanic — the majority traveled in third class at modest prices, while a small elite minority paid many times more for first-class accommodations.

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df.loc[df['survived']==0, 'fare'], label='Did Not Survive', shade=True)
sns.kdeplot(df.loc[df['survived']==1, 'fare'], label='Survived', shade=True)
plt.title('Fare Distribution by Survival Status')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.legend()
plt.show()
```

# Discussion for Task 3
---

The overlaid KDEs show that survivors’ fares cluster slightly higher than those who did not survive. Both groups share the same right-skew, but the survival curve’s density is stronger at higher fare values, while non-survivors dominate the low-fare range. This pattern suggests that higher-fare (and thus wealthier or higher-class) passengers had better survival chances, aligning with the historical accounts of evacuation favoring first- and second-class travelers.

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='pclass', y='fare', hue='sex')
plt.title('Fare Distribution by Passenger Class and Sex')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.legend(title='Sex')
plt.show()
```

# Discussion for Task 4
-----

Fares decrease steadily from 1st → 3rd class, confirming the expected class hierarchy.
Within each class, male passengers generally paid less than female passengers, though the median difference is small compared with the jump between classes.
The long upper whiskers and several high-value outliers highlight that a few first-class travelers paid dramatically more, emphasizing extreme economic inequality among passengers.

```python
plt.figure(figsize=(7,5))
sns.violinplot(data=df, x='sex', y='age', inner='box', palette='Set2')
plt.title('Age Distribution by Sex')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```

# Discussion for Task 5
---


Both sexes show roughly similar age spreads, but female passengers’ distribution is slightly younger on average and a bit more concentrated around early adulthood (20–35 yrs).
Males display a flatter, broader shape, with more passengers in middle age and older brackets.
This suggests the ship carried more adult men traveling independently and more young women traveling with family, consistent with historical demographics.

```python
plt.figure(figsize=(8,5))
sns.violinplot(data=df, x='sex', y='age', inner=None, color='lightgray')
sns.stripplot(data=df, x='sex', y='age', jitter=True, alpha=0.5)
plt.title('Age Distribution with Individual Observations')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```

# Discussion for Task 6
-----

Adding individual points reveals dense clusters around the 20–40 age range for both sexes and makes rare extremes (children under 5 and seniors over 70) immediately visible.
While the violin alone summarized shape, the strip overlay exposes actual sample density and outlier presence, showing that the apparent smooth tails correspond to just a handful of passengers.
This view bridges summary and detail, making it clear where the data are thick or thin rather than relying solely on smoothed distributions.
